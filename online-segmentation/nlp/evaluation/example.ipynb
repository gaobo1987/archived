```python
# How to run evaluations of a language model
import os
import sys
sys.path.append(os.path.abspath(os.path.join('..')))
sys.path.append(os.path.abspath(os.path.join('..', '..')))
from nlp.evaluation import utils 
lang = 'el'
gold_data = 'gold.json'
eval_data = 'eval.json'
```

```python
# ===========================================================================
# Step 1: Data preparation: 
# prepare the dataset that contains both ground-truth (gold) and predictions 
# (in this case, predicted by our online-segmenter)
# for evaluations in the next step, 
# if you already have such a dataset, you can skip this step
# ===========================================================================
from nlp.evaluation.utils import compose_evaluation_data
gold_data_path = os.path.abspath(os.path.join('data', lang, gold_data))
save_path = os.path.abspath(os.path.join('data', lang, eval_data))
compose_evaluation_data(gold_data_path, save_path, lang)
```

```python
# Note that Step 2, 3 and 4 can be done in parallel
```

```python
# ===========================================================================
# Step 2: Evaluation - POS
# ===========================================================================
from nlp.evaluation.utils import evaluate
eval_data_path = os.path.abspath(os.path.join('data', lang, eval_data))
output = evaluate(eval_data_path, key='pos')
output
```

```python
# ===========================================================================
# Step 3: Evaluation - Lemma
# ===========================================================================
from nlp.evaluation.utils import evaluate
eval_data_path = os.path.abspath(os.path.join('data', lang, eval_data))
output = evaluate(eval_data_path, key='lemma')
output
```

```python
# ===========================================================================
# Step 4: Evaluation - NER
# ===========================================================================
from nlp.evaluation.utils import evaluate
eval_data_path = os.path.abspath(os.path.join('data', lang, eval_data))
output = evaluate(eval_data_path, key='ner')
output
```

```python
gold = load_json('data/el/gold.json')
```

```python
def has_ner(arr):
    for i in arr:
        if 'ner' in i:
            return True
    return False
```

```python
gold_ner = []
for q in gold:
    if has_ner(q['gold']):
        gold_ner.append(q)
```
